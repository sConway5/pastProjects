```python
#Q2.1
coffee1 = reuters.fileids(categories = 'coffee')
coffeedoc = coffee1[0]
print(coffeedoc)
```

```python
#q2.2
fdist = nltk.FreqDist(w.lower() for w in reuters.words(coffeedoc))
fdist
```

```python
#q2.3
fdist.plot(7)
```

```python
#q3.1
tea = reuters.fileids(categories=['tea'])
coffee = reuters.fileids(categories=['coffee'])
coffee_and_tea = list(set(coffee).intersection(tea))
len(coffee_and_tea)
```

```python
#q3.2
corn = reuters.fileids(categories=['corn'])
wheat = reuters.fileids(categories=['wheat'])
wheat_and_corn = list(set(wheat).intersection(corn))
len(wheat_and_corn)
```

```python
#Q4.1
words1 = reuters.words(coffeedoc)
print(words1)
```

```python
#Q4.2
lowercase = []
for word in words1:
    lowercase.append(word.lower())
```
